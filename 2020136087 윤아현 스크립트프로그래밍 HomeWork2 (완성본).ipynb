#문제 2번.
경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.

```python
s = input("경로를 입력하세요 ! :")

def div_path(s) :
    
    I = []
    
    count = 0
    size = 0
    
    for i in range(len(s)) :
        if s[i] == "/" :
            count += 1
        else :
            continue
        
    for j in range(len(s)) :
        
        if s[j] == "/" :
            count -= 1
            
        if count < 1 :
            break
            
        size += 1
    
    a = s[ 0 : size+1]
    b = s[size+1 : ]
    
    I.append(a)
    I.append(b)
    
    return I

print(div_path(s))
```

변수 s에 경로에 해당하는 문자열을 입력 받는다.

div_path(s)함수를 만든다

디렉토리와 파일을 구분하기 위해서는 마지막 슬래시 '/'의 위치를 알아야 한다.
  문자열에서 '/'의 문자가 나올 때마다 count를 1씩 증가시켜 슬래시가 몇 개 있는지 알아낸다.
  
그 다음 마지막 슬래시 전과 후로 문자열의 길이가 각 몇으로 나누어지는지 알아야 한다.
  마지막 슬래시가 나오기 전까지 size의 길이를 1씩 증가시키고 마지막 슬래시가 나오면 반복문을 종료시킨다.
  
문자열 슬라이싱을 사용해 변수 a와 b에 사이즈의 길이를 기준으로 전과 후로 각각 저장한다.

append()를 사용하여 리스트에 a, b를 저장한다. 

#문제 3번.
두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.


```python
def list_union(lista, listb) :
    I = []
    
    for i in lista :
            I.append(i)
            
    for j in listb :
        if j not in I :
            I.append(j)
    
    I.sort()
    
    return I

print(list_union([1, 2, 3], [1, 2, 4]))
print(list_union([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_union([0, 1, 2], [0, 1, 2]))    
        
```

list_union함수

함수 안에 빈 리스트 I를 만든다.
lista의 요소들을 리스트 I에 다 추가한다.
만일 I에 요소가 없다면, append() 함수를 사용해 요소를 추가한다.
listb의 요소들을 확인한다. 만일 리스트 I에 중복된 요소가 없다면, 요소를 추가한다.
리스트 안에 있는 요소들을 순서대로 정렬하기 위해 sort를 사용한다.

#문제 4번.
두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '교집합'을 반환하는 함수 list_intersection(lista, listb)와 
'차집합'을 반환하는 함수 list_difference(lista, listb)를 작성하시오.


```python
def list_intersection(lista, listb) :
    listc = []
    for i in lista :
        if i in listb :
            listc.append(i)
    
    listc.sort()
    return listc

def list_difference(lista, listb) :
    
    listd = list_intersection(lista, listb)
    AB = []
    BA = []
    
    for i in lista :
        if i not in listd :
            if i not in AB :
                AB.append(i)
    
    for j in listb :
        if j not in listd :
            if j not in BA :
                BA.append(j)
    
    AB.sort()
    BA.sort()
    
    return AB, BA

print("교집합")
print(list_intersection([1, 2, 3], [1, 2, 4]))
print(list_intersection([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_intersection([0, 1, 2], [0, 1, 2]))
print("차집합[A-B , B-A]")
print(list_difference([1, 2, 3], [1, 2, 4]))
print(list_difference([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_difference([0, 1, 2], [0, 1, 2]))  
```

list_intersection() 함수

교집합을 나타낼 빈 리스트, listc를 만든다.
lista에 있는 요소들을 확인한다. 만일 listb에 이 요소들이 있을 경우 append() 함수를 사용해 listc에 요소를 추가한다.
리스트 안에 있는 요소들을 순서대로 정렬하기 위해 sort를 사용한다.
listc를 반환한다.

list_difference() 함수

차집합을 나타내기 위해서는 교집합의 요소들을 알아야하기 때문에 list_intersection함수를 실행하여 listd에 저장한다.
차집합 (A-B)와 (B-A)를 나타낼 리스트들을 각 AB, BA로 설정한다. (빈리스트로 초기화함.)
A-B : lista에 있는 요소들을 확인한다. 만일 listd에 이 요소들이 없을 경우 append() 함수를 사용해 AB에 요소를 추가한다.
중복으로 요소가 들어가는 것을 막기 위해 AB에도 이미 요소가 존재하는지 확인한다.
B-A : listb에 있는 요소들을 확인한다. 만일 listd에 이 요소들이 없을 경우 append() 함수를 사용해 BA에 요소를 추가한다.
중복으로 요소가 들어가는 것을 막기 위해 BA에도 이미 요소가 존재하는지 확인한다.
sort를 사용해 요소들을 정렬하고 AB, BA 모두 반환한다.

#문제 5번.
두 개의 양의 정수를 인자로 받아서 해당 범위안에 있는 소수(prime number)의 리스트를 출력하는 함수 print_primenumber(a, b)를 작성하시오.


```python
def print_primenumber(a, b):
    prime_list = []
    for i in range(a, b+1) :
        prime = True
        for j in range(2, i) :
            if i % j == 0 :
                prime = False
                break
        if prime :
            prime_list.append(i)
            
    a = "{0} 이상 {1} 이하 정수 중 소수 리스트 : {2}" .format(a, b, prime_list)
    print(a)
    
print_primenumber(5, 100)
print_primenumber(2, 10)
```

print_primenumber() 함수

소수들을 저장할 리스트 prime_list를 생성한다.(빈 리스트로 초기화한다.)
prime 변수를 사용하여 소수인지 확인을 한다.
a부터 b까지 반복하면서 이 두 정수 사이에 존재하는 숫자가 약수가 있는지 확인을 한다.
정수가 아무 수로도 나누어지지 않는다면 prime_list에 요소로 추가한다.

순수 텍스트를 변수에 저장한 뒤, 문자열에 존재한 특수문자를 re.sub()를 사용하여 삭제한다. 공백을 기준으로 list로 변경한 뒤 list의 길이를 센다.

```python
#link 2 문제
import requests

req = requests.get('https://blog.naver.com/ahyeon24')

source = req.text

print(source)
```

```python
#link 2 코드

import requests
import re

req = requests.get('https://blog.naver.com/ahyeon24')

source = req.text

def Checkblock(link) :
    
    check = False
    
    for i in range(len(link)) :
        
        if link[i:i+7] == '<title>' :
            check = True
            break
            
        else :
            continue
            
    return check


def Countblock(link) :
    
    count = 0
    num = 0
    
    for i in range(0, len(link)) : 
        if link[i:i+7] == '<title>'  :
            num = i+7
            for j in range(i, len(link)) : 
                if link[j:j+8] != '</title>'  :
                    count += 1
                    continue

                elif link[j:j+8] == '</title>' :
                    break
            break
        else :
            continue
            
    link = link[num: num+count-7] 
    
    return link

    
def Deleteblock(link) :
    
    HTML = Countblock(link)
    
    while Checkblock(HTML) == True :
        
        HTML = Countblock(HTML)
        
        continue
    
    return HTML

HTML = Deleteblock(source)
print(HTML)
result = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', HTML)        
list_result = result.split()
print("특수문자 제외 단어 수 : ", len(list_result))
```
